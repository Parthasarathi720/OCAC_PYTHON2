```python
import numpy as np
import matplotlib.pyplot as plt

company = ['Butter', 'chocolate', 'chicken', 'Buttermilk']
revenue = [45, 150, 90, 30]
profit = [50, 5, 87, 15]

xpos = np.arange(len(company))

plt.bar(xpos - 0.2, revenue, width=0.4, label="Revenue")
plt.bar(xpos + 0.2, profit, width=0.4, label="Profit")

plt.xticks(xpos, company)
plt.ylabel("Revenue (Bln)")
plt.title('Reliance Fresh')
plt.legend()
plt.savefig('plot2.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt

offlineretail = ['Butter', 'chocolate', 'chicken', 'Buttermilk']
Q1_profits = [35, 45, 100, 70]
Q2_profits = [38, 40, 105, 65]
Q3_profits = [30, 42, 120, 72]
Q4_profits = [25, 35, 115, 60]

plt.figure(1, figsize=(10, 10))

plt.subplot(221)
plt.bar(offlineretail, Q1_profits)
plt.title('Quarter1 profit')

plt.subplot(222)
plt.bar(offlineretail, Q2_profits)
plt.title('Quarter2 profit')

plt.subplot(223)
plt.bar(offlineretail, Q3_profits)
plt.title('Quarter3 profit')

plt.subplot(224)
plt.bar(offlineretail, Q4_profits)
plt.title('Quarter4 profit')

plt.suptitle('Profit on Quarter Basis')

plt.savefig('plot.jpg')

plt.show()
```
